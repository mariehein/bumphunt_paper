# Bump Hunt Plotting for CWoLa

```python
import bumphunt_plotting_utils as bp
import numpy as np
import matplotlib.pyplot as plt
import bumphunt_plotting_utils_no_labels as bpl
```

```python
general_directory = "/hpcwork/zu992399/bumphunt_paper/"
general_directory_project = "/work/rwth0934/bumphunt_paper/"
general_directory_work = "/work/zu992399/bumphunt_paper/"
plotting_direc = "plots/sig_formula/"

min_error = -0.5
max_error = 2.2

min_sig = -1
max_sig = 10
max_sig_no = 5.5
```

## Baseline Dataset

```python
folder = general_directory+"cwola_herwig_without_2/"
results, rel_results, true_results, rel_error, exp = bp.bump_hunt(folder)
cwola_herwig_error = np.mean(rel_error,axis=(1,2))

bp.plotting_error(rel_error, "cwola_without_herwig", min=min_error, max=max_error)
cwola_herwig_stat = np.mean(exp,axis=(1,2))
cwola_herwig_sys = np.std(rel_error, axis=(1,2))
print("alpha =",cwola_herwig_error)
print("alpha_std =", np.std(rel_error, axis=(1,2)))
print("sys_err/stat_err =",cwola_herwig_stat)
```

```python
folder = general_directory+"cwola_without/"
results, rel_results, true_results, rel_error, exp, info = bp.bump_hunt(folder, reuse=True)
cwola_error = np.mean(rel_error,axis=(1,2))

bp.plotting_error(rel_error, "cwola_without", min=min_error, max=max_error)
cwola_stat = np.mean(exp,axis=(1,2))
cwola_sys = np.std(rel_error, axis=(1,2))
print("alpha =", cwola_error)
print("alpha_std =", np.std(rel_error, axis=(1,2)))
print("stat =", cwola_stat)

results, rel_results, true_results, rel_error, exp = bp.bump_hunt_reuse(info, err=cwola_herwig_error)
bp.plotting(rel_results, true_results, "cwola_without_pythia_herwig", min=min_sig, max = max_sig_no)

err = np.array([max([cwola_herwig_sys[i],cwola_herwig_error[i]]) for i in range(3)])
results, rel_results, true_results, rel_error, exp = bp.bump_hunt_reuse(info, err=cwola_herwig_error, err_err=err)
bp.plotting(rel_results, true_results, "cwola_without_pythia_herwig", min=min_sig, max = max_sig_no)
```

```python
folder = general_directory+"cwola_norm/"
results, rel_results, true_results, rel_error, exp, info = bp.bump_hunt(folder, err=cwola_herwig_error, reuse=True)
bp.plotting(rel_results, true_results, "cwola_norm_pythia_herwig", min=min_sig, max = max_sig)
```

```python
folder = general_directory+"cwola_baseline_norm_randsignal/"
results, rel_results, true_results, rel_error, exp, info = bpl.bump_hunt(folder, err=cwola_herwig_error, reuse=True)
bpl.plotting(rel_results, true_results, "cwola_norm_pythia_herwig", min=min_sig, max = max_sig)
```
