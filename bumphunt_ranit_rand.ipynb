```python
import bumphunt_plotting_utils as bp
import numpy as np
import matplotlib.pyplot as plt
import bumphunt_plotting_utils_no_labels as bpl
import matplotlib.cm as col
```

```python
general_directory = "/hpcwork/zu992399/bumphunt_paper/"
general_directory_project = "/work/rwth0934/bumphunt_paper/"
general_directory_work = "/work/zu992399/bumphunt_paper/"
plotting = "plots/cathode/"

min_error = -0.5
max_error = 2.2

min_sig = -1
max_sig = 14
max_sig_no = 5.5
```

## Baseline without Signal

```python
folder = general_directory+"cathode_without_Herwig_ranit_rand/"
results, rel_results, true_results, rel_error, exp = bp.bump_hunt(folder)
ranit_Herwig_error = np.mean(rel_error,axis=(1,2))
cathode_stat = np.mean(exp,axis=(1,2))
cathode_sys_Herwig = np.std(rel_error, axis=(1,2))
print("alpha =",ranit_Herwig_error)
print("alpha_std =", np.std(rel_error, axis=(1,2)))
print("stat_error =",cathode_stat)
bp.plotting_error(rel_error, "cathode_Herwig", min=min_error, max=max_error, plotting_directory=plotting)
```

```python
folder = general_directory_project+"cathode_without_SB_stats_ranit_rand/"
results, rel_results, true_results, rel_error, exp = bp.bump_hunt(folder)
ranit_SB_error = np.mean(rel_error,axis=(1,2))
cathode_stat = np.mean(exp,axis=(1,2))
cathode_sys_SB = np.std(rel_error, axis=(1,2))
print("alpha =",ranit_SB_error)
print("alpha_std =", np.std(rel_error, axis=(1,2)))
print("stat_error =",cathode_stat)
bp.plotting_error(rel_error, "cathode_SB_without", min=min_error, max=max_error, plotting_directory=plotting)
```

```python
folder = general_directory_project+"cathode_without_ranit_rand/"
results, rel_results, true_results, rel_error, exp, info = bp.bump_hunt(folder, reuse=True)
cathode_error = np.mean(rel_error,axis=(1,2))
cathode_stat = np.mean(exp,axis=(1,2))
cathode_sys = np.std(rel_error, axis=(1,2))
print("alpha =",cathode_error)
print("alpha_std =", np.std(rel_error, axis=(1,2)))
print("stat_error =",cathode_stat)
bp.plotting_error(rel_error, "cathode_without", min=min_error, max=max_error, plotting_directory =plotting)

results, rel_results, true_results, rel_error, exp = bp.bump_hunt_reuse(info, err=ranit_SB_error)
bp.plotting(rel_results, true_results, "ranit_without_SB", min=min_sig, max = max_sig_no, plotting_directory=plotting)

results, rel_results, true_results, rel_error, exp = bp.bump_hunt_reuse(info, err=ranit_Herwig_error)
bp.plotting(rel_results, true_results, "ranit_without_Herwig", min=min_sig, max = max_sig_no, plotting_directory=plotting)

err = np.array([max([ranit_SB_error[i],ranit_Herwig_error[i]]) for i in range(3)])
results, rel_results, true_results, rel_error, exp = bp.bump_hunt_reuse(info, err=err)
bp.plotting(rel_results, true_results, "ranit_without_Herwig_SB", min=min_sig, max = max_sig_no, plotting_directory=plotting)
```
