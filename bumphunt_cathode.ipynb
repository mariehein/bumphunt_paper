```python
import bumphunt_plotting_utils as bp
import numpy as np
import matplotlib.pyplot as plt
```

```python
general_directory = "/hpcwork/zu992399/bumphunt_paper/"
general_directory_project = "/work/rwth0934/bumphunt_paper/"
general_directory_work = "/work/zu992399/bumphunt_paper/"

min_error = -0.5
max_error = 2.2

min_sig = -1
max_sig = 10
max_sig_no = 5.5
```

## Baseline without signal

```python
folder = general_directory_work+"cathode_without_Herwig_randDE_avg_2/"
results, rel_results, true_results, rel_error, exp = bp.bump_hunt(folder)
cathode_herwig_error = np.mean(rel_error,axis=(1,2))
bp.plotting_error(rel_error, "cathode_without_herwig", min=min_error, max=max_error)
cathode_herwig_stat = np.mean(exp,axis=(1,2))
cathode_herwig_sys = np.std(rel_error, axis=(1,2))
print("alpha =",cathode_herwig_error)
print("alpha_std =", np.std(rel_error, axis=(1,2)))
print("stat_error =",cathode_herwig_stat)
```

```python
folder = general_directory+"cathode_without_SB_randDE/"
results, rel_results, true_results, rel_error, exp = bp.bump_hunt(folder)
cathode_SB_error = np.mean(rel_error,axis=(1,2))
bp.plotting_error(rel_error, "cathode_without_SB", min=min_error, max=max_error)
cathode_SB_stat = np.mean(exp,axis=(1,2))
cathode_SB_sys = np.std(rel_error, axis=(1,2))
print("alpha =",cathode_SB_error)
print("alpha_std =", np.std(rel_error, axis=(1,2)))
print("stat_error =",cathode_SB_stat)
```

```python
folder = general_directory+"cathode_without_randDE_avg/"
results, rel_results, true_results, rel_error, exp, info = bp.bump_hunt(folder, reuse=True)

cathode_error = np.mean(rel_error,axis=(1,2))
bp.plotting_error(rel_error, "cathode_without", min=min_error, max=max_error)
cathode_stat = np.mean(exp,axis=(1,2))
cathode_sys = np.std(rel_error, axis=(1,2))
print("alpha =",cathode_error)
print("alpha_std =", np.std(rel_error, axis=(1,2)))
print("stat_error =",cathode_stat)

results, rel_results, true_results, rel_error, exp = bp.bump_hunt_reuse(info, err=cathode_SB_stats_error)
bp.plotting(rel_results, true_results, "cathode_without_SB_stats", min=min_sig, max = max_sig_no)
```
